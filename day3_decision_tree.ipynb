```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.tree import DecisionTreeRegressor,export_text
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data",header=None,sep="\s+")
```

```python
df.head()
```

```python
df.columns = ['mpg',
'cylinders',
'displacement',
'horsepower',
'weight',
'acceleration',
'model',
'origin',
'car'
]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['car'] = le.fit_transform(df['car'])
```

```python
df['horsepower']=df['horsepower'].str.replace("?","0")
```

```python
df['horsepower'] = df['horsepower'].astype(float)
```

```python
X = df.drop('mpg',axis=1)
Y= df['mpg']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=10)
```

```python
model = DecisionTreeRegressor()
model.fit(X_train,Y_train)
Y_pred = model.predict(X_test)
```

```python
print(mean_squared_error(Y_test,Y_pred))
```

```python
print(r2_score(Y_test,Y_pred)
```

```python
print(r2_score(Y_test,Y_pred))
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
%matplotlib inline
```

```python
X, y = load_breast_cancer(return_X_y=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
clf = DecisionTreeClassifier()
```

```python
path = clf.cost_complexity_pruning_path(X_train, y_train)
```

```python
help(clf.cost_complexity_pruning_path(X_train, y_train))
```

```python
path
```

```python
ccp_alphas, impurities = path.ccp_alphas, path.impurities
```

```python
plt.figure(figsize=(10, 6))
plt.plot(ccp_alphas, impurities)
plt.xlabel("effective alpha")
plt.ylabel("total impurity of leaves")
```

```python
clfs = []
for ccp_alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    clf.fit(X_train, y_train)
    clfs.append(clf)
```

```python
tree_depths = [clf.tree_.max_depth for clf in clfs]
plt.figure(figsize=(10,  6))
plt.plot(ccp_alphas[:-1], tree_depths[:-1])
plt.xlabel("effective alpha")
plt.ylabel("total depth")
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc_scores = [accuracy_score(y_test, clf.predict(X_test)) for clf in clfs]
```

```python
tree_depths = [clf.tree_.max_depth for clf in clfs]
plt.figure(figsize=(10,  6))
plt.grid()
plt.plot(ccp_alphas[:-1], acc_scores[:-1])
plt.xlabel("effective alpha")
plt.ylabel("Accuracy scores")
```

```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Prakhar-Katiyar/CloudyML/main/Decision%20Tree%20Pruning/Iris.csv")

df.drop(columns=['Id'], inplace=True)
```

```python
#Dividing dataset into features(X) and label(y)
X = df.drop(columns=['Species'])
y = df['Species']
print(X.shape)
print(y.shape)
```

```python
#Splitting dataset to train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=45,stratify=y)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0)
clf.fit(X_train,y_train)
y_train_pred = clf.predict(X_train)
y_test_pred = clf.predict(X_test)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(19,12))
features = df.columns
classes = ['Iris-setosa','Iris-versicolor', 'Iris-virginica']
plot_tree(clf,feature_names=features,class_names=classes,filled=True)
plt.show()
```
